**Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose ?**

**Answer - ** Simple Linear Regression (SLR) is a statistical method used to model the relationship between two variables ‚Äî one independent variable (X) and one dependent variable (Y) ‚Äî by fitting a straight line to the observed data.

Simple Linear Regression estimates the relationship between X and Y using the equation:

 $y=\beta _{0}+\beta _{1}x+\epsilon ¬†$

Where:

- Y ‚Üí Dependent (response) variable

 - X ‚Üí Independent (predictor) variable

 - $\beta _{0}$ ‚Üí Intercept (value of Y when X = 0)

 - $\beta _{1}$ ‚Üí Slope (change in Y for a one-unit change in X)

 - $\epsilon $ ‚Üí Error term (difference between observed and predicted values)

üîπ Purpose of Simple Linear Regression:

**Prediction:**

 - To predict the value of the dependent variable (Y) based on a known value of the independent variable (X).

Example: Predicting house price based on its size.

 - Understanding Relationships:

 - To determine if and how strongly two variables are related.

Example: Studying how advertising budget (X) affects sales revenue (Y).

 - Trend Analysis:

To identify and quantify trends over time (e.g., sales growth per year).

**Example:**

Suppose you want to predict a student‚Äôs exam score (Y) based on the number of study hours (X).

After fitting a regression line, you might get:

**Score=30+5√óHours**

Interpretation:

 - The intercept (30) means even with 0 study hours, the predicted score is 30.

 - The slope (5) means for each additional hour studied, the score increases by 5 points.

**Question 2: What are the key assumptions of Simple Linear Regression?**

**Answer -**

**1. Linearity**

This assumption states that the relationship between the independent variable (\(x\)) and the dependent variable (\(y\)) is linear.

- **What it means:** For every one-unit increase in the independent variable, the dependent variable should increase or decrease by a constant amount.

- **How to check:** Visually inspect a scatter plot of the data. The points should appear to fall along a straight line.

- **What happens if violated:** The model will fail to capture the true trend, leading to systematically biased predictions that are too high or too low for certain ranges of the independent variable.

**2. Independence of errors**

This assumption means that the residuals (the differences between the observed and predicted values) are independent of each other.

- **What it means:** The error for one observation is not influenced by the error of any other observation. This is especially important for time series data, where consecutive data points may be correlated.

- **How to check:** Examine a plot of residuals versus the predicted values or the order of data collection. There should be no discernible pattern. The Durbin-Watson test is a formal statistical test for this.

- **What happens if violated:** Correlated errors lead to underestimated standard errors and unreliable significance tests.

**3. Homoscedasticity (Constant variance)**

This assumes that the variance of the residuals is constant across all levels of the independent variable.

- **What it means:** The spread of the residuals around the regression line should be roughly the same for all values of \(x\). The opposite, heteroscedasticity, is when the variance of the residuals changes as \(x\) changes, often in a fan or cone shape.

- **How to check:**
 Create a plot of residuals versus fitted (predicted) values. The points should be randomly scattered in a uniform band centered at zero.

 - **What happens if violated:** Heteroscedasticity can result in less precise and inefficient parameter estimates, leading to biased standard errors and invalid hypothesis tests.

**4. Normality of residuals**¬†This assumption states that the residuals of the model are normally distributed.

- **What it means:** When plotted on a histogram, the residuals should form a bell-shaped curve. A normal distribution of errors is important for creating valid confidence intervals and conducting hypothesis tests.

- **How to check:** Create a histogram or a Normal Q-Q plot of the residuals. A Q-Q plot of normally distributed residuals will closely follow a straight diagonal line.

- **What happens if violated:** While the parameter estimates themselves may still be unbiased, the standard errors and p-values will be less reliable, especially with smaller sample sizes.

**Question 5: What is the method of least squares in linear regression?**

**Answer**-The method of least squares in linear regression is a statistical technique used to determine the line of best fit for a given set of data points by minimizing the sum of the squared differences (called residuals) between the observed values and the predicted values on that line.

 - **Define the error: For a straight-line equation**

 **$y=a+bx$**, the error for each data point $(x_{i},y_{i})$ is the difference between the observed value $y_{i}$ and the value predicted by the line $({y_{i}}=a+bx_{i})$.

 - **Square the errors:**

  The difference $(y_{i}-\^{y_{i}})$ is squared to ensure that all errors are positive and to give more weight to larger errors.

 - **Sum the squared errors:**

  The squared errors for all data points are added together to get the total sum of squares, also known as the cost function.

 - **Minimize the sum:**

  Calculus is used to find the specific values for the slope $(b)$ and the y-intercept $(a)$ that will minimize this total sum of squared errors.

**Question 6: What is Logistic Regression? How does it differ from Linear Regression?**

Logistic regression is a statistical method used for classification problems that predicts the probability of a categorical outcome. Unlike linear regression, which forecasts a continuous value, logistic regression uses a logistic or sigmoid function to output a value between 0 and 1, representing the probability of a given class or event.

**Key characteristics of Logistic Regression:**

 - **Purpose:**

 Primarily used for classification tasks. For example, predicting if an email is spam (yes/no) or if a patient has a certain disease (present/absent).

 - **Mathematical function:**

  It uses a sigmoid function, also known as the logistic function, to map any real-valued number into a probability between 0 and 1. The output is an S-shaped curve, which represents the probability of the outcome.

 - **Output interpretation:**

  The output is a probability score between 0 and 1. A threshold (e.g., 0.5) is used to convert this probability into a definitive class prediction. For example, a probability over 0.5 might be classified as "yes".


 - **Estimation method:**

  It uses Maximum Likelihood Estimation (MLE) to find the coefficients that maximize the likelihood of observing the given data.

**Question 7: Name and briefly describe three common evaluation metrics for regression
models.**

**Answer -**  Three common evaluation metrics for regression models are Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared$(R^{2})$. These metrics help assess how well a model's predictions align with the actual observed values.

**1. Mean Squared Error (MSE)**

MSE is the average of the squared differences between the predicted values and the actual values.

Formula: $MSE=\frac{1}{n}\sum _{i=1}^{n}(y_{i}-\^{y}_{i})^{2}$

 - $(y_{i})$ is the actual value.

 - $({\^y}_{i})$ is the predicted value.

 - $(n)$ is the number of data points.

 **Key characteristics:**

 - Since it squares the errors, larger errors are penalized more heavily than smaller ones, making it sensitive to outliers.

  - The resulting value is not in the same unit as the target variable, which can make it less intuitive to interpret.

  - A lower MSE indicates a better model fit.

  **2. Root Mean Squared Error (RMSE)**

RMSE is the square root of the MSE and is one of the most widely used metrics for regression.

**Formula:**

 $(RMSE=\sqrt{\frac{1}{n}\sum _{i=1}^{n}(y_{i}-\^{y}_{i})^{2}})$

 **Key characteristics:**

 - It measures the magnitude of the error in the same units as the target variable, making it more interpretable than MSE.

 - Like MSE, it heavily penalizes large errors, as the errors are squared before being averaged.

 - A value of 0 indicates a perfect fit, with lower values indicating a better model.

 **3. R-squared $(R^{2})$**

   $R^{2}$ is a statistical measure that represents the proportion of the variance for a dependent variable that is explained by the independent variables in a regression model. It is also known as the coefficient of determination.

   **Formula:**
   
   $R^{2}=1-\frac{\text{Residual\ Sum\ of\ Squares\ (RSS)}}{\text{Total\ Sum\ of\ Squares\ (TSS)}}$

   **Key characteristics:**
   
   - The value ranges from 0 to 1.
   
   - $(R^{2}=1)$ means the model perfectly explains the variance of the target variable.

   - $R^{2}=0$ means the model explains none of the variance.

   - A higher $R^{2}$ is generally better, but it can be misleading. Adding more independent variables will always increase the $R^{2}$, even if the new variables do not improve the model's predictive power. For this reason, the adjusted $R^{2}$ is often used for multiple regression.
   
   - It doesn't indicate if a model is biased, so it should be interpreted alongside other metrics and diagnostic plots.

**Question 8: What is the purpose of the R-squared metric in regression analysis?**

**Answer -** The purpose of the R-squared ($R^{2}$) metric in regression analysis is to measure the proportion of the variation in the dependent variable that can be explained by the independent variable(s) in the model. In simpler terms, it's a "goodness-of-fit" measure that indicates how well the regression model fits the observed data.

**Interpretation and use of R-squared**

A value from 0 to 1: R-squared is a value between 0 and 1, often expressed as a percentage.

- **R-squared of 0:**

 Means the model explains none of the variability of the dependent variable around its mean. In this case, predicting the dependent variable using the model is no better than simply using the mean of the dependent variable.

- **R-squared of 1:**

 Represents a perfect fit, where the model explains all the variability in the dependent variable. This is rarely seen with real-world data.

- **Value between 0 and 1:**

For example, an R-squared of 0.70 (or 70%) means the model can account for 70% of the variance in the dependent variable.

**Limitations of R-squared**

While a higher R-squared value generally indicates a better model fit, it is not always a reliable indicator of model quality on its own.

- **Risk of overfitting:**

 R-squared will always increase or stay the same when you add more independent variables to a model, even if those variables are not actually significant. This can lead to overfitting, where a model fits the training data well but fails to generalize to new, unseen data.

- **No indication of bias:**

A high R-squared doesn't necessarily mean the predictions are unbiased. It is always important to check residual plots to confirm the model's assumptions hold true and that the errors are random and unbiased.

- **Context is key:**

What constitutes a "good" R-squared value depends heavily on the field of study. In some social sciences, an R-squared of 0.5 might be considered strong, while in physical sciences, a much higher value might be expected.


**Question 9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)**

```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Example dataset (X = independent variable, y = dependent variable)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for sklearn (2D input)
y = np.array([2, 4, 5, 4, 5])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print the slope (coefficient) and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)

# Predicting values (optional)
y_pred = model.predict(X)
print("Predicted values:", y_pred)
```

**Question 10: How do you interpret the coefficients in a simple linear regression model?**

In a simple linear regression model, the relationship between a dependent variable (\(Y\)) and a single independent variable (\(X\)) is represented by a straight line, described by the equation:

   $Y=\beta _{0}+\beta _{1}X+\epsilon $

The interpretation of the coefficients involves understanding the meaning of  $(\beta _{0})$ and $(\beta _{1})$

**Interpretation of the Coefficients:**

- **Intercept $(\beta _{0})$**

  - The expected value of Y when X = 0.

  - It tells you where the regression line crosses the Y-axis.

 Example: if $(\beta _{0})$ = 2 ,it means when X= 0,
Y is predicted to be 2.

**Note:** Sometimes X=0 is not meaningful (e.g., years of experience = 0), so interpret the intercept carefully depending on context.


**Slope $(\beta _{1})$**


- Represents the change in Y for a one-unit increase in X, keeping all else constant.

- Example: If  $\beta _{1}$ = 0.8
it means that for every 1-unit increase in X, Y increases by 0.8 units on average.

- If $\beta _{1}$  is positive, the relationship is direct (Y increases as X increases).

- If $(\beta _{1})$ is negative, the relationship is inverse (Y decreases as X increases).

**Example:**

Suppose your regression equation is:

  Y=2+0.5X

 - Intercept (2): When X=0, the predicted Y = 2

 - Slope (0.5): For each 1-unit increase in X,Y increases by 0.5 units.
